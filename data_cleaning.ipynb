Feature Engineering and Encoding

```python
country_to_continent = {
    1: 'Oceania', 2: 'Europe', 3: 'Europe', 4: 'North America', 5: 'North America',
    6: 'Oceania', 7: 'Europe', 8: 'Europe', 9: 'Europe', 10: 'Europe',
    11: 'Europe', 12: 'Unknown', 13: 'Europe', 14: 'Europe', 15: 'Europe',
    16: 'Europe', 17: 'Europe', 18: 'Europe', 19: 'Europe', 20: 'Asia',
    21: 'Europe', 22: 'Europe', 23: 'Europe', 24: 'Europe', 25: 'Europe',
    26: 'North America', 27: 'Europe', 28: 'Europe', 29: 'Europe', 30: 'Europe',
    31: 'Europe', 32: 'Europe', 33: 'Europe', 34: 'Europe', 35: 'Europe',
    36: 'Europe', 37: 'Europe', 38: 'Europe', 39: 'Europe', 40: 'Asia',
    41: 'Europe', 42: 'North America', 43: 'Unknown', 44: 'Unknown',
    45: 'Unknown', 46: 'Unknown', 47: 'Unknown'
}
```

```python
class DataPreprocessor:
    def __init__(self, country_to_continent):
        # Store the country-to-continent mapping as an attribute
        self.country_to_continent = country_to_continent

    def map_country_to_continent(self, df):
        # Map country to continent using the provided mapping
        df['continent'] = df['country'].map(self.country_to_continent)
        return df

    def drop_unknown_continents(self, df):
        # Drop rows where the continent is 'Unknown'
        df = df[df['continent'] != 'Unknown']
        return df

    def drop_country_column(self, df):
        # Drop the 'country' column as it's no longer needed
        df = df.drop(columns=['country'])
        return df

    def split_clothing_model(self, df):
        # Split the 'page2_clothing_model' into type and number
        df['clothing_model_number'] = df['page2_clothing_model'].str[1:].astype(int)
        df = df.drop(columns=['page2_clothing_model'])
        return df

    def encode_main_category(self, df):
        # Encode 'page1_main_category' into descriptive categories
        df['page1_main_category'] = df['page1_main_category'].map({
            1: 'trousers', 2: 'skirts', 3: 'blouses', 4: 'sale'
        })
        return df

    def encode_price_2(self, df):
        # Encode the 'price_2' column
        df['price_2'] = df['price_2'].map({1: 1, 2: 0})
        return df

    def preprocess_data(self, df):
        # Apply all preprocessing steps
        df = self.map_country_to_continent(df)
        df = self.drop_unknown_continents(df)
        df = self.drop_country_column(df)
        df = self.split_clothing_model(df)
        df = self.encode_main_category(df)
        df = self.encode_price_2(df)
        return df
```

```python
# Create an instance of the DataPreprocessor class
preprocessor = DataPreprocessor(country_to_continent) 

# Apply preprocessing to the train and test DataFrames
train_df = preprocessor.preprocess_data(train_df)
test_df = preprocessor.preprocess_data(test_df)
```

```python
# Session-based metrics functions
class SessionMetrics:
    def __init__(self, train_df, test_df):
        self.train_df = train_df
        self.test_df = test_df
    # Session Length: Total number of clicks per session
    def compute_session_length(self, df):
        session_metrics = df.groupby('session_id')['order'].max().reset_index()
        session_metrics.rename(columns={'order': 'session_length'}, inplace=True)
        return session_metrics
    
    # Computes the number of clicks per product category per session
    def compute_category_clicks(self, df):
        category_clicks = df.groupby(['session_id', 'page1_main_category']).size().unstack(fill_value=0)
        category_clicks.columns = [f"clicks_{col}" for col in category_clicks.columns]
        category_clicks.reset_index(inplace=True)
        return category_clicks
    
    # Computes the browsing path (sequence of categories visited in each session)
    def compute_browsing_path(self, df):
        browsing_path = df.groupby('session_id')['page1_main_category'].apply(lambda x: ' → '.join(x)).reset_index()
        browsing_path.rename(columns={'page1_main_category': 'browsing_path'}, inplace=True)
        return browsing_path
    
    # Computes the first click (first category visited in each session)
    def compute_first_click(self, df):
        first_click = df.groupby('session_id')['page1_main_category'].first().reset_index()
        first_click.rename(columns={'page1_main_category': 'first_click'}, inplace=True)
        return first_click
    
    # Computes the first click (first category visited in each session)
    def compute_last_click(self, df):
        last_click = df.groupby('session_id')['page1_main_category'].last().reset_index()
        last_click.rename(columns={'page1_main_category': 'last_click'}, inplace=True)
        return last_click
    
    # Computes all the session-based metrics and merges them into one DataFrame
    def compute_all_metrics(self, df):
        session_data = self.compute_session_length(df)
        category_clicks = self.compute_category_clicks(df)
        browsing_path = self.compute_browsing_path(df)
        first_click = self.compute_first_click(df)
        last_click = self.compute_last_click(df)

        # Merging all metrics
        session_data = session_data.merge(category_clicks, on='session_id', how='left')
        session_data = session_data.merge(browsing_path, on='session_id', how='left')
        session_data = session_data.merge(first_click, on='session_id', how='left')
        session_data = session_data.merge(last_click, on='session_id', how='left')

        return session_data

    def get_train_metrics(self):
        # Calculate metrics for the training data
        return self.compute_all_metrics(self.train_df)

    def get_test_metrics(self):
        # Calculate metrics for the testing data
        return self.compute_all_metrics(self.test_df)
```

```python
# Create an instance of the SessionMetrics class
session_metrics_obj = SessionMetrics(train_df, test_df)

# Compute metrics for the training dataset
train_metrics = session_metrics_obj.get_train_metrics()

# Compute metrics for the testing dataset
test_metrics = session_metrics_obj.get_test_metrics()

# Now you can merge the metrics with the original train_df and test_df
train_df = train_df.merge(train_metrics, on='session_id', how='left')
test_df = test_df.merge(test_metrics, on='session_id', how='left')

# Print the updated DataFrames
print("Updated Training Data with Metrics:")
print(train_df)

print("Updated Testing Data with Metrics:")
print(test_df)
```

One-Hot Encoding for categorical features


```python
categorical_columns = ['page1_main_category', 'continent']

def encode_data(df):
    return pd.get_dummies(df, columns=categorical_columns)

train_df = encode_data(train_df)
test_df = encode_data(test_df)

# Align train and test data to ensure same feature columns
train_df , test_df = train_df.align(test_df , join = 'left', axis=1 , fill_value= 0)
```

```python
# Logic for "purchased_or_not"
def determine_purchase(row):
    # If price is higher than average, less likely to purchase
    if row["price_2"] == 1:
        return 0
    # Check if significant clicks and main category engagement
    if (
        (row["page1_main_category_sale"] and row["clicks_sale"] > 1) or
        (row["page1_main_category_skirts"] and row["clicks_skirts"] > 1) or
        (row["page1_main_category_trousers"] and row["clicks_trousers"] > 1)
    ):
        return 1
    # If session length is long (e.g., > 20), assume purchase intent
    if row["session_length"] > 4:
        return  1
    # Otherwise, no purchase
    return 0
```

```python
# Apply the logic
train_df["purchased_or_not"] = train_df.apply(determine_purchase, axis=1)
# Apply the logic
test_df["purchased_or_not"] = test_df.apply(determine_purchase, axis=1)
```
